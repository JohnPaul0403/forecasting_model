# Gathering Data for the Model

In this notebook we will gather and fecth raw data from the market via Yahoo Finance for the model. For this part we will fecth the last 10 years for daily data for the SPY and the VIX (Implied Volatility Index). Both stocks data will give clearly benchmarks. The decision for the daily lyes under the certainity of the movement and vast data availibility from it, also models such as Garch and TFT can learn better from this timeframe rather than the noisy smaller dataframes. 

## Data fetching

For this case we will use Yfinance API to fetch the data of SPY and vix, we will fetch a total worth of 13 years, since we will train the model with 10 years of historical data and use 3 years for testing the model also for the VIX we will only get 3 years worth of data with the only use being to becnhmark with the model predictions.

```python
import numpy as np
import pandas as pd
import yfinance as yf
```

```python
spy = yf.download('SPY', start='2012-01-01', interval='1d')
spy.head()
```

```python
vix = yf.download('^VIX', start='2022-01-01', interval='1d')
vix.head()
```

## Data Sanity Check
